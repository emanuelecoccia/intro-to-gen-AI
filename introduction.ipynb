# Welcome

Welcome to the "Introduction to genAI course". 

This course is going to give you an overview of what you can build with an LLM and some creativity.

# Program

This is the outline of the course:

1) In the first chapter, you are going to create a small program that connects to OpenAI and lets the user chat with an LLM. 
2) In the second chapter, you will create a User Interface, so that the user can chat from there instead of the terminal.
3) In the third chapter, you will parse a corpus of documents, separate it into chunks, embed them and store them in a file.
4) In the fourth chapter, you will modify the initial program to retrieve the most significant document from the vector file for each of the user questions. You will then pass these contexts to the LLM, to get informed answers.
5) In the fifth chapter, you will replace the vector file with an actual vector database.
6) In the sixth chapter, you will learn how to connect to the internet to retrieve factual information for your LLM.

After this, you will have a good grip on genAI.
